```python
import twint
'''To solve; RuntimeError: This event loop is already running'''
import nest_asyncio
nest_asyncio.apply()

# Configure
c = twint.Config()

# Parameter setting
c.Limit = 200
c.Search = '"서울 버스"'
c.Since = '2015-01-01'
c.Output = 'seoul_bus_tweet.json'
c.Popular_tweets = True
c.Store_json = True
c.Hide_output = True
```

```python
import pandas as pd

raw_df = pd.read_json('seoul_bus_tweet.json', lines=True)

# tweet_df = raw_df.loc[:, ['date', 'time', 'username', 'tweet', 'retweets_count', 'likes_count']]
tweet_df = raw_df.iloc[:, [3, 4, 7, 10, 16, 17]]
tweet_df
```

```python
# 텍스트에서 한글만 추출
from soynlp.normalizer import *
# 띄어쓰기 정리
from pykospacing import spacing
```

```python
for i in range(len(tweet_df)):
    modified_str = only_hangle(tweet_df.iloc[i, 3])
    modified_str = emoticon_normalize(modified_str)
    tweet_df.iloc[i, 3] = spacing(modified_str)
    
tweet_df['tweet']
```
