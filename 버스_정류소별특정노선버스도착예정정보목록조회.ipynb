## 연습코드

```python
import requests
import urllib.parse as urlparse
import json
import pandas as pd
import numpy as np

# URL 만들기 함수
def get_request_query(operation1, operation2, params, serviceKey):
    URL = "http://openapi.tago.go.kr/openapi/service"
    params = urlparse.urlencode(params)
    request_query = URL + '/' + operation1 + '/' + operation2 + '?' + params + '&' + 'ServiceKey' + '=' + serviceKey
    return request_query


# 요청 URL과 오퍼레이션
URL = "http://openapi.tago.go.kr/openapi/service"

OPERATION1 = "ArvlInfoInqireService"
OPERATION2 = "getSttnAcctoSpcifyRouteBusArvlPrearngeInfoList"  

# 서비스키
SERVICEKEY = "Ee5WLqN4iRCKuFUsxlAF1P9anyOX5vH%2BOFG2%2BYM%2BcEoNQOg9emMEyyKM37eAmmVnl1ZxgTalHHL90VNl1B1zlg%3D%3D"

routeid_bus_stop5_name = pd.read_csv("버스_정류소아이디_정보.csv") # 노선 id랑 정류소id 여기서 가져옴

#데이터 (변수이름 정하기)
a = pd.DataFrame(columns = ( "노선번호(routeno)", "노선아이디(routeid)", 
                                        "정류소아이디(nodeid)", "정류소명(nodenm)",
                                        "도착예상시간[초](arrtime)", "남은정류장수(arrprevstationcnt)")) 

#routeno, routeid, nodeid, nodenm, arrtime, arrprevstationcnt

# 파라미터, 노선id는 버스_주요정류소_정보.csv에서 가저온다

cityCode  = 34010
nodeId = routeid_bus_stop5_name["정류소아이디(nodeid)"]  #정류소ID 
routeId = routeid_bus_stop5_name["노선아이디(routeid)"] #노선ID 
numOfRows = 10
_type = "json"

bus_route_id_idx = 0 # api에서 노선id별 정류소id들을 구하기 위해 사용
index = 0
for _ in range(len(nodeId)):
    if(index == 1): # 테스트 1개만
        break
    
    
    try:
        PARAMS = {'cityCode':cityCode, 'routeId':routeId[index], 'nodeId': nodeId[index], 'numOfRows': numOfRows, '_type': _type}

        # URL만들기
        request_query = get_request_query(OPERATION1, OPERATION2, PARAMS, SERVICEKEY)
        print('request_query:', request_query)

        # #GET
        response = requests.get(url = request_query)

        #상태 확인
        #print('status_code:' + str(response.status_code))

        r_dict = json.loads(response.text)
        r_response = r_dict.get("response")
        r_body = r_response.get("body")
        r_items = r_body.get("items")
        r_item = r_items.get("item")
        #print(type(r_item))
    except:
        r_header = r_response.get("header")
        #"resultMsg":"NORMAL SERVICE."
        if(r_header["resultMsg"] == "NORMAL SERVICE."):
            # 정류소에 버스 도착 데이터 없음, 
            print("데이터 없음") # 표시 어떻게 할지 고민하기
        else:
            print("api 문제: " + str(r_header["resultMsg"]))
            
    try:
        for item in r_item: #원하는 정류소id만 가져오기 위한 for문
            routeno = item.get("routeno")
            routeid = item.get("routeid")
            nodeid = item.get("nodeid")
            nodenm = item.get("nodenm")
            arrtime = item.get("arrtime")
            arrprevstationcnt = item.get("arrprevstationcnt")

            a.loc[bus_route_id_idx] = [ routeno, routeid, nodeid, nodenm, arrtime, arrprevstationcnt ]


            bus_route_id_idx += 1

    except:
            routeno = r_item["routeno"]
            routeid = r_item["routeid"]
            nodeid = r_item["nodeid"]
            nodenm = r_item["nodenm"]
            arrtime = r_item["arrtime"]
            arrprevstationcnt = r_item["arrprevstationcnt"]

            a.loc[bus_route_id_idx] = [ routeno, routeid, nodeid, nodenm, arrtime, arrprevstationcnt ]
            bus_route_id_idx += 1

    index+=1

```
